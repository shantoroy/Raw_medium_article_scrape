```python
from matplotlib.pyplot import figure
plt.rcParams["font.family"] = "Times New Roman"
figure(num=None, figsize=(12, 7))

with open("../conference_medium_data/author_data/related_data_rm_duplicacy.json","r") as f:
    data = json.load(f)


count19 = 0
count18 = 0
count17 = 0
count16 = 0
count15 = 0
count14 = 0

#l = data_sheet.sort_values(by='CreationDate')

month2019 = [0] * 13
month2018 = [0] * 13
month2017 = [0] * 13
month2016 = [0] * 13
month2015 = [0] * 13
month2014 = [0] * 13


for key in data:
    date = str(key["publish_date"])
#for k in data_sheet_Questions.publish_date:
#   date, time = str(k).split(" ")  # Date and time splitting
# date parsing
    y, m, d = date.split("-")
    if y == "2019":
        m = int(m)
        month2019[m] += 1
        count19 += 1

    elif y == "2018":
        m = int(m)
        month2018[m] += 1
        count18 += 1

    elif y == "2017":
        m = int(m)
        month2017[m] += 1
        count17 += 1

    elif y == "2016":
        m = int(m)
        month2016[m] += 1
        count16 += 1

    elif y == "2015":
        m = int(m)
        month2015[m] += 1
        count15 += 1

    elif y == "2014":
        m = int(m)
        month2014[m] += 1
        count14 += 1

y_pos = np.arange(60)
years = ["January 15", "January 16", "January 17", "January 18", "January 19"]
f = plt.figure()
#plt.set_ylim([0, 300])
plt.bar(y_pos,month2015[1:]+month2016[1:]+month2017[1:]+month2018[1:]+month2019[1:])
# width = 0.25
plt.xticks(range(1,60,12), years)
plt.ylabel('No of Posts')
plt.title("Number of posts over the time")
#plt.show()
plt.savefig('medium_posts_over_time.pdf', dpi=300)
```
