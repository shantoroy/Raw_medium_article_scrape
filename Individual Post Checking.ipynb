```python
import time
from bs4 import BeautifulSoup
import requests
import json
import re
```

```python
class PostDetais(object):
    def __init__(self, soup, link=None):
        self.page_soup = soup
        self.link = link

    def get_title(self):
        #class_names = ['jz b ka kh ay',
                       #'graf graf--h3 graf--leading graf--title',
                       #'graf graf--h3 graf-after--figure graf--trailing graf--title']
        try:
            for my_tag in self.page_soup.find_all('h1'):
                title = my_tag.text
                return title
        except Exception as e:
            error_trace = {}
            error_trace["link"] = self.link
            error_trace["method"] = "get_title"
            error_trace["message"] = str(e)
            print(json.dumps(error_trace, indent=4))
        return ""

    def get_author_name(self):
        # class by a = ["bb bc bd be bf bg bh bi bj bk if bn bo bp bq"]
        class_names = ["cf eh ei ch be kb kc bb bc bd ay",
                       "ao cs ed aq by kq ef eg kr ei ea",
                       "ao cs ed aq by ki ef eg kj ei ea", 
                       "ao cs ed aq by la ef eg lb ei ea",
                       "ao cz ej aq by kl ce cf cg ch eg",
                       "ao cr dx aq bx kl dz ea km ec du",
                       "ao cz ej aq by kt ce cf cg ch eg",
                       "br dt du bt iz kd ke kf kg kh fo",
                       "ao cs ed aq by ks ef eg kt ei ea",
                       "ao cs ed aq by lc ef eg ld ei ea",
                       "br dt du bt jg jh ji jj jk jl fo",
                       "ao cr dx aq bx kb dz ea kc ec du",
                       "ar ea eb bf av kb ax ay az ba cr",
                       "br dt du bt iz jz ka kb kc kd fo",
                       "br dt du bt iz kd ke kf kg kh fo",
                       "ao cs ec aq by kr ee ef ks eh dz"]
        try:
            for my_tag in self.page_soup.find_all('span',{'class': class_names}):
                name = my_tag.text
                return name
        except Exception as e:
            error_trace = {}
            error_trace["link"] = self.link
            error_trace["method"] = "get_title"
            error_trace["message"] = str(e)
            print(json.dumps(error_trace, indent=4))
        return ""

    def get_date(self):
        #  class by a = ["bb bc bd be bf bg bh bi bj bk if bn bo bp bq"]
        class_names = ["cf eh ei ch be kb kc bb bc bd ck", 
                       "ao cs ed aq by kq ef eg kr ei at",
                       "ao cs ed aq by ki ef eg kj ei at",
                       "ao cs ed aq by la ef eg lb ei at",
                       "ao cz ej aq by kl ce cf cg ch at",
                       "ao cr dx aq bx kl dz ea km ec at",
                       "ao cz ej aq by kt ce cf cg ch at",
                       "br dt du bt iz kd ke kf kg kh bw",
                       "ao cs ed aq by ks ef eg kt ei at",
                       "ao cs ed aq by lc ef eg ld ei at",
                       "br dt du bt jg jh ji jj jk jl bw",
                       "ao cr dx aq bx kb dz ea kc ec at",
                       "ar ea eb bf av kb ax ay az ba bi",
                       "br dt du bt iz jz ka kb kc kd bw",
                       "br dt du bt iz kd ke kf kg kh bw",
                       "ao cs ec aq by kr ee ef ks eh at"]
        try:
            for my_tag in self.page_soup.find_all('span',{'class': class_names}):
                date_time = my_tag.text
                return date_time
        except Exception as e:
            error_trace = {}
            error_trace["link"] = self.link
            error_trace["method"] = "get_title"
            error_trace["message"] = str(e)
            print(json.dumps(error_trace, indent=4))
        return ""

    def get_read(self):
        try:
            for my_tag in self.page_soup.find_all(class_="readingTime"):
                read = my_tag.get('title')
                return read
        except Exception as e:
            error_trace = {}
            error_trace["link"] = self.link
            error_trace["method"] = "get_title"
            error_trace["message"] = str(e)
            print(json.dumps(error_trace, indent=4))
        return ""

    def get_upvote(self):
        class_names = ["bp bq br bs bt bu bv bw bx by bz ca cb cc cd ce",
                       "cy cz ax ay az ba bb bc bd be du dv bh bi dw dx",
                       "df dg ax ay az ba bb bc bd be eb ec bh bi ed ee",
                       "cw cx ax ay az ba bb bc bd be cy cz bh bi da db",
                       "bb bc bd be bf bg bh bi bj bk bl bm bn bo bp bq",
                       "dr ds bm bn bo bp bq br bs bt dt du bw bx dv dw",
                       "bb bc bd be bf bg bh bi bj bk bl bm bn bo bp bq"]
        try:
            for my_tag in self.page_soup.find_all('button', {
                'class': class_names}):
                upvotes = my_tag.text
                return upvotes
        except Exception as e:
            error_trace = {}
            error_trace["link"] = self.link
            error_trace["method"] = "get_upvote"
            error_trace["message"] = str(e)
            print(json.dumps(error_trace, indent=4))
        return ""

    def get_body(self):
        news_body = ""
        for paragraphs in self.page_soup.find_all(
                class_='graf graf--p graf-after--p'):
            news_body += paragraphs.text.rstrip().lstrip()
        return news_body

    def get_post_content(self):
        try:
            # class_="section-content"
            # we are using <article> tag as section-content class no longer works
            # everything is inside the article tag including title & author names
            for content in self.page_soup.find_all('article'):
                return content.text
        except Exception as e:
            error_trace = {}
            error_trace["link"] = self.link
            error_trace["method"] = "get_post_content"
            error_trace["message"] = str(e)
            print(json.dumps(error_trace, indent=4))
        return ""

    def get_response(self):
        class_names = ["cn", "av", "bz"]
        try:
            for my_tag in self.page_soup.find_all('span', {
                'class': class_names}):
                res = my_tag.text
                return res
        except Exception as e:
            error_trace = {}
            error_trace["link"] = self.link
            error_trace["method"] = "get_response"
            error_trace["message"] = str(e)
            print(json.dumps(error_trace, indent=4))
        return ""
    
    def test_json_response(self):
        try:
            for my_tag in self.page_soup.find_all('script', {
                'type': "application/ld+json"}):
                res = my_tag.text
                return res
        except Exception as e:
            error_trace = {}
            error_trace["link"] = self.link
            error_trace["method"] = "get_response"
            error_trace["message"] = str(e)
            print(json.dumps(error_trace, indent=4))
        return ""
    
    def test2_json_response(self):
        try:
            # source: https://stackoverflow.com/questions/13323976/how-to-extract-a-json-object-that-was-defined-in-a-html-page-javascript-block-us 
            #script = re.compile('window.__APOLLO_STATE__ = ({.*})', re.DOTALL)
            #json_text = script.search(str(self.page_soup)
            #script = tree.xpath('//script[contains(., "window.__APOLLO_STATE__")]/text()')[0]
            #json_text = re.search(self.page_soup"window.__APOLLO_STATE__ = (.*?);$", script).group(1)
            for tag in self.page_soup.find_all('script'):
                if 'window.__APOLLO_STATE__' in tag.text:
                    data = tag.text
            return data[26:]
        except Exception as e:
            error_trace = {}
            error_trace["link"] = self.link
            error_trace["method"] = "get_response"
            error_trace["message"] = str(e)
            print(json.dumps(error_trace, indent=4))
        return ""
```
